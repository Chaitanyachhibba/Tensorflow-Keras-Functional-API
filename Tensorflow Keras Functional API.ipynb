```python
from tensorflow.keras.utils import plot_model
```

```python
first_input = Input(shape=(2, ))
first_dense = Dense(1, name='first_dense'  )(first_input)
model = Model(inputs=[first_input], outputs=first_dense)
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
second_input = Input(shape=(2, ))
second_dense = Dense(1, name='second_dense' )(second_input)
model = Model(inputs=[second_input], outputs=second_dense)
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
merge_one = concatenate([first_dense, second_dense],name='merge_one')
model = Model(inputs=[first_input,second_input], outputs=merge_one)
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
third_input = Input(shape=(1, ))
merge_two = concatenate([merge_one, third_input],name='merge_two')
model= Model(inputs =[first_input,second_input,third_input], outputs= merge_two,name='MODEL')
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
first_input = Input(shape=(2, ))
first_dense = Dense(1, name='first_dense' )(first_input)

second_input = Input(shape=(2, ))
second_dense = Dense(1, name='second_dense' )(second_input)

merge_one = concatenate([first_dense, second_dense],name='merge_one')

third_input = Input(shape=(1, ))
merge_two = concatenate([merge_one, third_input],name='merge_two')

model = Model(inputs=[first_input, second_input, third_input], outputs=merge_two,name='MODEL')
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
ip = Input(shape=(10, ))
X1 = Dense(7, name='X1'  )(ip)
X2 = Dense(5, name='X2' )(X1)
X3 = Dense(3, name='X3' )(X2)
X4 = Dense(4, name='X4' )(X2)
X5 = concatenate([X3,X4], name = 'X5' )
model = Model( inputs = [ip], outputs = X5, name = 'M1')
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```
